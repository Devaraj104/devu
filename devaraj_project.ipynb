```python
url = "https://raw.githubusercontent.com/Naresh312143/mahalingpur-python-ml/refs/heads/main/Indian%20Unicorn%20startups%202023%20updated.csv"
df = pd.read_csv(url)
cols = df.columns.tolist()
cols = [col.strip() for col in cols]
cols = [col.replace(' ', '_') for col in cols]
cols = [col.replace('(', '') for col in cols]
cols = [col.replace(')', '') for col in cols]
cols = [col.replace('^', '') for col in cols]
cols = [col.replace('*', '') for col in cols]
df.columns = cols

df['Entry'] = pd.to_datetime(df['Entry'], errors='coerce')
df['Primary_City'] = df['Location'].apply(lambda x: str(x).split('/')[0] if pd.notnull(x) else 'Unknown')
entry_valuation_col = None
for col in df.columns:
    if 'Entry_Valuation' in col:
        entry_valuation_col = col
        break

if entry_valuation_col is not None:
    df['Valuation_Growth'] = df['Valuation_$B'] / df[entry_valuation_col]
else:
    print("Error: 'Entry_Valuation' column not found after cleaning.")

df['No_of_Investors'] = df['Select_Investors'].apply(lambda x: len(str(x).split(',')) if pd.notnull(x) else 0)
```
